```python

%pip install geopandas matplotlib

import geopandas as gpd
import matplotlib.pyplot as plt
from urllib.request import urlopen
import json

# Load Chile shapefile with high resolution (10m instead of 110m)
# 10m provides much finer detail than 110m
url = "https://naciscdn.org/naturalearth/10m/cultural/ne_10m_admin_0_countries.zip"

try:
    # Load directly from URL with high resolution
    world = gpd.read_file(url)
    chile = world[world['NAME'] == 'Chile']
    
    # Create figure with higher DPI and resolution for better quality
    fig, ax = plt.subplots(figsize=(10, 16), dpi=300)
    chile.plot(ax=ax, color='#4A90D9', edgecolor='#2E5C8A', linewidth=0.5)
    
    # Remove axes
    ax.axis('off')
    
    # Create directory if it doesn't exist
    import os
    output_dir = r'C:\Users\esteb\GitHub\Proyecto-Final-Muestreo\04_Informe\Informe_Final\Imagenes'
    os.makedirs(output_dir, exist_ok=True)
    
    # Save as vector format (PDF or SVG) with high quality
    plt.savefig(os.path.join(output_dir, 'mapa_chile.svg'), 
                format='svg', bbox_inches='tight', transparent=True, dpi=300)
    plt.savefig(os.path.join(output_dir, 'mapa_chile.pdf'), 
                format='pdf', bbox_inches='tight', transparent=True, dpi=300)
    
    print("✓ Mapa de alta resolución (10m) guardado exitosamente en SVG y PDF")
    print(f"  Resolución: 300 DPI")
    print(f"  Tamaño: {fig.get_figwidth()} x {fig.get_figheight()} pulgadas")
    plt.show()
    
except Exception as e:
    print(f"Error: {e}")
    print("Intentando alternativa con archivo local...")
```
